```
import random
import cProfile
```

```
mod=10**9+7
def solve(a,b):
    if b==0:
        return 1
    if b==1:
        return a
    if b%2==0:
        return ((solve(a,b//2)%mod)**2)%mod
    return ((a%mod)*(((solve(a,b//2)%mod)**2)%mod))%mod
```

```
def post_order(arr, l, r):
    if l > r:
        return
    else:
        mid = (r+l) // 2
        post_order(arr, l, mid - 1)
        post_order(arr, mid + 1, r)
        #print(a[mid])
```

```
def partition(arr, l, r):

    x = arr[r]
    i = l
    for j in range(l, r):

        if arr[j] >= x:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1

    arr[i], arr[r] = arr[r], arr[i]
    return i


def kthLargest(arr, l, r, k):

    # if k is smaller than number of
    # elements in array
    if (k > 0 and k <= r - l + 1):

        # Partition the array around last
        # element and get position of pivot
        # element in sorted array
        index = partition(arr, l, r)

        # if position is same as k
        if (index - l == k - 1):
            return arr[index]

        # If position is more, recur
        # for left subarray
        if (index - l > k - 1):
            return kthLargest(arr, l, index - 1, k)

        # Else recur for right subarray
        return kthLargest(arr, index + 1, r,
                          k - index + l - 1)
    return -1
```
